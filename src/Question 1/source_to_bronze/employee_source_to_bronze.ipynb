```python
base_path = "/Volumes/assignment/default/employee_vol/input/"

employee_df = spark.read.csv(base_path + "employee.csv", header=True, inferSchema=True)
department_df = spark.read.csv(base_path + "department.csv", header=True, inferSchema=True)
country_df = spark.read.csv(base_path + "country.csv", header=True, inferSchema=True)
```

```python
bronze_path = "/Volumes/assignment/default/employee_vol/source_to_bronze/"

employee_df.write.mode("overwrite").csv(bronze_path + "employee")
department_df.write.mode("overwrite").csv(bronze_path + "department")
country_df.write.mode("overwrite").csv(bronze_path + "country")
```
